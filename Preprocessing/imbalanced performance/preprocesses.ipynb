```python
def getRowIndexOfStepIndex(data):
    stepStartEnd = []
    first = 0
    stepindex = 1
    for i in range(len(data['Data_processed']['Data']['Step_Index'])):
        if(data['Data_processed']['Data']['Step_Index'][i]!=stepindex):
            stepStartEnd.append([stepindex, first, i-1])
            first = i
            stepindex +=1
    stepStartEnd.append([stepindex, first, len(data['Data_processed']['Data']['Step_Index'])-1])
    return stepStartEnd

def getTimeInSteps(data, stepStartEnd):
    timeframes = []
    for i in range(len(stepStartEnd)):
        timeframes.append([i+1, data['Data_processed']['Data']['Test_Times'][stepStartEnd[i][2]]-data['Data_processed']['Data']['Test_Times'][stepStartEnd[i][1]]])
    return timeframes

def getVoltageInSteps(data, stepStartEnd):
    VoltageV = []
    for i in range(len(stepStartEnd)):
        VoltageV.append([i+1, data['Data_processed']['Data']['VoltageV'][stepStartEnd[i][1]], data['Data_processed']['Data']['VoltageV'][stepStartEnd[i][2]]])
    return VoltageV

def getTempInSteps(data, stepStartEnd):
    temp1 = []
    temp2 = []
    temp3 = []
    temp4 = []
    tempA = []
    for i in range(len(stepStartEnd)):
        temp1.append([i+1, data['Data_processed']['Data']['TemperatureC_Cell_1'][stepStartEnd[i][1]], data['Data_processed']['Data']['TemperatureC_Cell_1'][stepStartEnd[i][2]]])
        temp2.append([i+1, data['Data_processed']['Data']['TemperatureC_Cell_2'][stepStartEnd[i][1]], data['Data_processed']['Data']['TemperatureC_Cell_2'][stepStartEnd[i][2]]])
        temp3.append([i+1, data['Data_processed']['Data']['TemperatureC_Cell_3'][stepStartEnd[i][1]], data['Data_processed']['Data']['TemperatureC_Cell_3'][stepStartEnd[i][2]]])
        temp4.append([i+1, data['Data_processed']['Data']['TemperatureC_Cell_4'][stepStartEnd[i][1]], data['Data_processed']['Data']['TemperatureC_Cell_4'][stepStartEnd[i][2]]])
        tempA.append([i+1, data['Data_processed']['Data']['Ambient_TemperatureC'][stepStartEnd[i][1]], data['Data_processed']['Data']['Ambient_TemperatureC'][stepStartEnd[i][2]]]) 
    return temp1, temp2, temp3, temp4, tempA

def getCurrentAmpInSteps(data, stepStartEnd):
    Amp1 = []
    Amp2 = []
    Amp3 = []
    Amp4 = []
    currentA = []
    for i in range(len(stepStartEnd)):
        Amp1.append([i+1, data['Data_processed']['Data']['CurrentA_Cell_1'][stepStartEnd[i][1]], data['Data_processed']['Data']['CurrentA_Cell_1'][stepStartEnd[i][2]]])
        Amp2.append([i+1, data['Data_processed']['Data']['CurrentA_Cell_2'][stepStartEnd[i][1]], data['Data_processed']['Data']['CurrentA_Cell_2'][stepStartEnd[i][2]]])
        Amp3.append([i+1, data['Data_processed']['Data']['CurrentA_Cell_3'][stepStartEnd[i][1]], data['Data_processed']['Data']['CurrentA_Cell_3'][stepStartEnd[i][2]]])
        Amp4.append([i+1, data['Data_processed']['Data']['CurrentA_Cell_4'][stepStartEnd[i][1]], data['Data_processed']['Data']['CurrentA_Cell_4'][stepStartEnd[i][2]]])
        currentA.append([i+1, data['Data_processed']['Data']['CurrentA'][stepStartEnd[i][1]], data['Data_processed']['Data']['CurrentA'][stepStartEnd[i][2]]])
    return Amp1, Amp2, Amp3, Amp4, currentA

def getVoltageInStepsAvergae(data, stepStartEnd):
    VoltageVmean = []
    for i in range(len(stepStartEnd)):
        VoltageVmean.append(np.mean([data['Data_processed']['Data']['VoltageV'][stepStartEnd[i][1]:stepStartEnd[i][2]]]))
    return VoltageVmean

ex1 = getRowIndexOfStepIndex(M1_mixed_aged_0_10)
ex2 = getRowIndexOfStepIndex(M1_mixed_aged_0_40)
getVoltageInStepsAvergae(M1_mixed_aged_0_10, ex1)
getVoltageInStepsAvergae(M1_mixed_aged_0_40, ex2)

ex3 = HallSensorInSteps.getRowIndexOfStepIndexForCell(H1_calibration['Data_struct'])
ex4 = HallSensorInSteps.getDataForCellInStepIntevall(H1_calibration['Data_struct'], ex3, 'CurrentA')
print(ex4)
```
