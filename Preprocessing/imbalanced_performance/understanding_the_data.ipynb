```python
largeDataSet = IndividualProcess.getLongCellDataset()
cells = {
    key: group
    for key, group in largeDataSet.groupby("CellID")
}
for cell_id, cell_df in cells.items():
    dfs = {
        key: group
        for key, group in cell_df.groupby("StepIndex")
    }
    stats = {
        k: (v["CurrentData"].min(), v["CurrentData"].max())
        for k, v in dfs.items()
    }
    print(cell_id, stats)
```

```python
stats = (
    largeDataSet
        .groupby(["CellID", "StepIndex"])["CurrentData"]
        .agg(min_value="min", max_value="max")
        .reset_index()
)
with pd.option_context(
    "display.max_rows", None,
    "display.max_columns", None,
    "display.width", None,
    "display.max_colwidth", None,
):
    print(stats)
```
